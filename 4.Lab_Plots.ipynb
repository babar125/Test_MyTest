```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# A simple line plot

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.figure()
plt.plot(x,y,'ro')
fig=plt.figure()
plt.plot(x,y+1,':b')
fig.savefig('ab.png',dpi=300)
```

# Adding labels (horizontal and vertical) to the plot

```python
ax=plt.axes()
ax.plot(x,np.sin(x))
ax.set(xlim=(0,10),ylim=(-1,1),xlabel="x",ylabel='sin(x)',title='illustration');
```

# Scatter Plot
A scatter plot is used to show the relationship between two variables. 
There should be atleast two variables. If there are more than two variables then color/size can be used to represent the third dimension. It is used for continuous or numerical data.

```python
rng=np.random.RandomState(42)
x=rng.randn(100)
y=rng.randn(100)
colors=rng.randn(100)
sizes=1000*rng.randn(100)
plt.scatter(x,y,c=colors,s=sizes,cmap='viridis',alpha=0.5)
plt.colorbar()
```

# As we generated data points by using normal distribution with mean 0 and standard deviation of unity, majority of the points are located around 0. 

# Error Bars

Error bars can be used with line, dot, or bar plot in order to show the uncertainity in the measurement of the data points.

```python
x=np.linspace(0,10,50)
dy=0.8
y=np.sin(x)+dy*np.random.randn(50)
plt.errorbar(x,y,yerr=dy, fmt='o', color='green', ecolor='blue', elinewidth=3, capsize=4)
```

# Contour Plot
A contour plot is used to represent 3-dimensional data into a 2-dimensional plot. 

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contour(X, Y, Z, colors='black');
```

# In above figure the negative values of z are represented by dashed line and positive values are represented by the solid lines.

```python
plt.contour(X, Y, Z, 20, colors='black');
```

# here N=20 represents the contour level. As evident the number of contour lines are increased.

```python
X,Y # grid values obtained by using mesh grid
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
plt.colorbar()
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy');
plt.colorbar()
```

# In above figure, the positive values are represented by black color and negative values are represented by the reddish color. 

# Histogram
A histogram is used for visualizing the distribution of a variable in a dataset. The data is divivded in to a number of bins and the frequency of the data points falling within each bin is reported.

```python
data = np.random.randn(1000) # by deaful 10 bins are defined
plt.hist(data);
```

```python
plt.hist(data, bins=30, normed=True, alpha=0.5, histtype='stepfilled', color='steelblue', edgecolor='none');
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
count,bins_edges=np.histogram(x1,bins=40)# in order to varify the boundaries for bins and the respective frequency
```

# Two-dimenasional Histogram 

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x,y= np.random.multivariate_normal(mean, cov, 10000).T 
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
counts,xedges,yedges
```

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
cb = plt.colorbar(label='count in bin')
```

# subplots
within a figure, we can define a number of subplots.

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),fontsize=18, ha='center')
```

```python
#if all the subplots share the axis, then it is better the defines the axis tickmarks or labels just onces
# by activating the sharex/ sharey options.
fig, ax = plt.subplots(2, 3, sharex='col', sharey='row')
```

# To be noted, in above figure, the tick marks are shared by all the subplots, and are not repeatedly displayed for each subplot.

# Customizing subplots

```python
# Create some normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 3000).T
# Set up the axes with gridspec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=0.2, wspace=0.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)
# scatter points on the main axes
main_ax.plot(x, y, 'ok', markersize=3, alpha=0.2)
# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical', color='gray')
x_hist.invert_yaxis()
y_hist.hist(y, 40, histtype='stepfilled', orientation='horizontal', color='gray')
y_hist.invert_xaxis()
```

# In above figure by using markers alone, the plot methods is used for generating the scatter plot. The correlation between two variables is evident from the result.

# Annotation
It is used for adding details in to a plot.

```python
fig, ax = plt.subplots(facecolor='lightgray')
ax.axis([0, 10, 0, 10])
# transform=ax.transData is the default, but we'll specify it anyway
ax.text(1, 5, ". Data: (1, 5)", transform=ax.transData)
ax.text(0.5, 0.1, ". Axes: (0.5, 0.1)", transform=ax.transAxes)
ax.text(0.2, 0.2, ". Figure: (0.2, 0.2)", transform=fig.transFigure);
```

```python
fig, ax = plt.subplots()
x = np.linspace(0, 20, 1000)
ax.plot(x, np.cos(x))
ax.axis('equal')
ax.annotate('local maximum', xy=(6.28, 1), xytext=(10, 4), arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('local minimum', xy=(5 * np.pi, -1), xytext=(2, -6), arrowprops=dict(arrowstyle="->", connectionstyle="angle3,angleA=0,angleB=-90"));
```

# 3d-plots

```python
from mpl_toolkits import mplot3d
ax = plt.axes(projection='3d')
# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')
# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
